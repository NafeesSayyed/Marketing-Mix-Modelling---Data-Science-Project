```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mmm_processed_dataset.csv')
df = df.dropna(subset=['sales'])
```

```python
X = df.drop('sales', axis = 1)
y = df['sales']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
split = int(len(df) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train_scaled = sc_X.fit_transform(X_train)
X_test_scaled = sc_X.transform(X_test)
y_train_scaled = sc_y.fit_transform(y_train.values.reshape(-1,1)).ravel()
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf', C= 1.7, epsilon = 0)
regressor.fit(X_train_scaled, y_train_scaled)
```

```python
y_pred_scaled = regressor.predict(X_test_scaled)
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R^2: {r2 * 100:.2f}%")
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"MAPE: {mape * 100:.2f}%")
```
